```python
import numpy as np
import matplotlib.pyplot as plt

def linf(param,x):
    return param[0] + param[1]*x
```

```python
points_x = [1,2,3]
points_y = [1,4,-2]
grad_x = []
grad_y = []

plt.scatter(points_x,points_y)
plt.scatter(grad_x, grad_y)

for i in range(len(points_y)-1):
    grad_x.append((points_x[i+1] + points_x[i])/2)
    abl_y = (points_y[i+1] - points_y[i])/(points_x[i+1] - points_x[i])
    grad_y.append(abl_y)
    plt.plot(np.linspace(points_x[i],points_x[i+1],100),linf([points_y[i],abl_y],np.linspace(0,points_x[i+1]-points_x[i],100)))
```

```python
# Beispiel-Daten
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([10, 12, 14, 13, 15])

x2 = np.array([0, 1.5, 2.5, 3.5, 5])
y2 = np.array([9, 11, 13, 14, 16])

# Interpolation von y2 auf x1
y2_interp = np.interp(x1, x2, y2)

# Differenz berechnen
diff = y1 - y2_interp

plt.plot(x1,y1,marker='x')
plt.plot(x2,y2,marker='x')
plt.scatter(x1,y2_interp)
plt.scatter(x1,diff)
```
